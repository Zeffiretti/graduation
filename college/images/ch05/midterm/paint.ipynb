```python
# 创建一个点数为 8 x 6 的窗口, 并设置分辨率为 80像素/每英寸
plt.figure(figsize=(12, 4), dpi=300)

# 柱子总数
N = 8
# 包含每个柱子对应值的序列
values = (6, 5, 7, 6, 1, 10, 5, 6)

# 包含每个柱子下标的序列
index = np.arange(N)

# 柱子的宽度
width = 0.9

# 绘制柱状图, 每根柱子的颜色为紫罗兰色
bars = plt.barh(index, values, width, color="#E2A093")
for bar in bars:
    bar.set_edgecolor("k")
    bar.set_linewidth(0.8)
bars[4].set_color("#59DECE")
bars[4].set_edgecolor("k")
bars[4].set_linewidth(0.8)

# 设置横轴标签
plt.ylabel("$a_{t}$", fontsize=16)
# 设置纵轴标签
# plt.xlabel('cnt', fontsize=16)

# 添加标题
# plt.title('Monthly average rainfall')

# 添加纵横轴的刻度
plt.yticks(index, ('$a_{8}$', '$a_{7}$', '$a_{6}$', '$a_{5}$', '$a_{4}$', '$a_{3}$', '$a_{2}$', '$a_{1}$'), fontsize=16)
plt.xticks(np.arange(0, 11, 1), fontsize=16)

# 添加图例
# plt.legend(loc="upper right")

plt.savefig('actioncnt.png', bbox_inches='tight')
plt.show()
```
